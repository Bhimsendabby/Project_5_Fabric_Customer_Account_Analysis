#Importing Packages

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
from pyspark.sql.types import *
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType
```

## We have 8 columns in the csv file
![image_1771957004146.png](./image_1771957004146.png "image_1771957004146.png")

```python
# Define the schema explicitly
# Giving the names to the columns
schema = StructType([
    StructField("SalesOrderNumber", StringType(), True),
    StructField("SalesOrderLineItem", IntegerType(), True),
    StructField("OrderDate", DateType(), True),
    StructField("CustomerName", StringType(), True),
    StructField("EmailAddress", StringType(), True),
    StructField("ItemName", StringType(), True),
    StructField("Quantity", IntegerType(), True),
    StructField("UnitPrice", DoubleType(), True),
    StructField("TaxAmount", DoubleType(), True)
])
```

```python
# Creating a array which contain file year
year = ["2019", "2020", "2021"]
```

## Creating User Defined Function which copies the data from URL to volumn (defined location)

```python
import os

#User defined function

def copy(path):
    # 1. Download the file to the driver's local nodes
    url = f"https://raw.githubusercontent.com/MicrosoftLearning/mslearn-databricks/main/data/{path}.csv"
    os.system(f"wget {url} -O /Volumes/bhim_bricks/dbo/bhim_vol/githubdata/sales_{path}.csv")
    # dbutils.fs.cp("file:/tmp/sales_2019.csv", "/Volumes/bhim_bricks/dbo/bhim_vol/githubdata/sales_2019.csv")

for i in year:
    copy(i)
```
